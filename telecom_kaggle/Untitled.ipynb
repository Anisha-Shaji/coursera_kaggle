```python
df = df.drop(['Unnamed: 0','customerID'], axis = 'columns') 
```

```python
column_list = df.columns.values.tolist()
bin_col = []
not_bin_col=[]
tri_col = []
for i in column_list:
    #print(df[i],'---------')
    print(df[i].unique())
    if len(df[i].unique()) > 3:
        not_bin_col.append(i)
    elif len(df[i].unique())==3:
        tri_col.append(i)
    else:
        bin_col.append(i)
       
```

```python
 df=df.drop(['gender','PaperlessBilling'], axis = 'columns') 
```

```python
df.head()
```

```python
for i in bin_col:
    if i != 'SeniorCitizen':
        df[i] = df[i].map({'Yes': 1,'No':0})
    else:
        df[i]=df[i]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
g.figure.savefig("output_heatmap1.png")
g
```

```python
# to plot pdfs of all the variables
import matplotlib.pyplot as plt
import seaborn as sns
for column in df:
    plt.figure()
    sns.distplot(df[column],rug=True,hist=False, label='UW', kde_kws={'bw':0.1})
```

```python
df = pd.concat([df,pd.get_dummies(df['season'], prefix='season')],axis=1)
df.drop(['season'],axis=1, inplace=True)
```

```python
tri_col
```

```python
not_bin_col
```

```python
for i in tri_col:
    df = pd.concat([df,pd.get_dummies(df[i], prefix=i)],axis=1)
    df.drop([i],axis=1, inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
df=df.drop(['PaymentMethod'],axis=1)
```

```python
y = df[['Churn']]
X = df[df.columns.difference(['Churn'])]
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state= 9 )
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score

clf = LinearRegression(normalize=True)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)

print(r2_score(y_test,y_pred))
```
